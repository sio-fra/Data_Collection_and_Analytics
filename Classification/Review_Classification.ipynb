# Data Collection Functions

A function is defined below which uses the BeautifulSoup package to first extract the URL for each establishment under each selected category, then extract and compile all of the reviews for all of the establishments in that category and their corresponding star rating.

```python
#function to create dataframe with reviews and corresponding star ratings
def get_reviews(category):
    rev=list() #create list with reviews to append
    stars=list() #create list with star rating to append
    url_base='http://mlg.ucd.ie/modules/yalp/' #base of the urls required
    url_category_str='/%s_list.html' %category #to be added on to full url 
    full_category_url = url_base+url_category_str #create full url for that category
    html = urllib.request.urlopen(full_category_url).read() #read html code retrieved from calling the url
    soup = BeautifulSoup(html, 'html.parser') #use the BeautifulSoup package to parse the html code
    for line in soup.findAll('a'): #find all lines with <a in the html 
        extra_url=line.attrs['href'] #extract the 'href' which will give us the extra portion of the url to extract specific reviews per establishment
        get_review_url=url_base+extra_url #create full url for each establishment used to request reviews for that establishment
        html = urllib.request.urlopen(get_review_url).read() #read html code retrieved from calling the url
        soup = BeautifulSoup(html, 'html.parser') #use the BeautifulSoup package to parse the html code
        
        for star_rating in soup.findAll(['p'], class_=["rating"]): #find all lines with <p in the html with a class equal to 'rating'
            stars.append(star_rating.img['alt']) #extract the number of stars per review and append it onto the list of stars
            
        for review in soup.findAll(['p'],class_=["review-text"]):#find all lines with <p in the html with a class equal to 'review-text'
            rev.append(review.text) #append the text comment to the list of reviews
    
    stars=pd.Series(stars).str.extract('(\d+)').astype(int) #extract the number of stars only from the string and save number as an integer
    df=pd.DataFrame()
    df['review']=rev #create review column
    df['star_rating']=stars #create number of stars column
    return df
```

A function is defined below which creates a new column in an existing dataframe based on the star_rating column, assigning either a positive (+1) or negative (-1) value. A positive value is assigned if the reviewer left a star rating greater or equal to four and a negative value is assigned if the reviewer left a star rating less than four

```python
#function to add column to dataframe classifying the review as either positive or negative
def create_category_column(df, category):
    pos=0
    neg=0
    for i in range(len(df)):
        if df.loc[i, 'star_rating']>=4: #if there are 4 or more stars class the review as positive
            df.loc[i, 'class']=1
            pos+=1
        elif df.loc[i, 'star_rating']<4: #if there are less than 4 stars class the review as negative
            df.loc[i, 'class']=-1
            neg+=1
        else: 
            print('error in assigning class to review')
    perc_pos=(pos/len(df))*100
    perc_neg=(neg/len(df))*100
    print('for category: %s'%category)
    print('proportion of reviews that are classed as positive is %.1f percent' %perc_pos)
    print('proportion of reviews that are classed as negative is %.1f percent\n\n' %perc_neg)
    return df
```

# Data Preprocessing & Prediction Model Creation

The below function accepts tokenized text (i.e. the text has been converted to a list of its constituent words) and reduce the tokens to their stem format e.g jumped would convert to jump, faster would convert to fast, this will ensure words with a generally similar meanings are aggregated. 

```python
def lemmatize_tokens(tokens): #function to lemmatize tokens 
    lemmatizer = nltk.stem.WordNetLemmatizer() #create lemmatizer function object from the nltk package that will extract the stem word from word conjugates
    lemma_tokens = [] #initialise empty list of tokens 
    for token in tokens:
        lemma_tokens.append( lemmatizer.lemmatize(token) ) #append lemmatized token to the list of lemma token
    return lemma_tokens #return the list of lemmatized tokens
```

The below function converts text to a list of constituent words then convert these words it to their stem format using the above function.

```python

def lemma_tokenizer(text): #function to convert text to tokens then lemmatize
    standard_tokenizer = CountVectorizer().build_tokenizer() #create tokenizer from scikit CountVectorizer package 
    tokens = standard_tokenizer(text) #convert text to tokens
    lemma_tokens = lemmatize_tokens(tokens) #call function to lemmatize tokens 
    return lemma_tokens #return the list of lemmatized tokens    
```

The below function accepts a vectorizer and data fitted to the vectorizer then displays the ten most frequent terms from the vectorizer.

```python
def display_common_terms(X, vectorizer, category):
    terms = vectorizer.get_feature_names() #extract all feature names from the vectorizer
    freqs = X.sum(axis=0)# find the summed frequency of each term
    # sort the indexes of the array by value, and then reverse it
    sorted_term_indexes = freqs.argsort() #sort frequency values in ascending order
    sorted_term_indexes = sorted_term_indexes[0, ::-1] #swap to descending order
    print('The most frequent terms for the %s category are: ' %category)
    for i in range(10): #print the 10 most frequent terms
        term_index = sorted_term_indexes[0,i]
        print("%s = %.0f" % ( terms[term_index], freqs[0,term_index]))
    print('\n\n')
```

The below function accepts a dataframe and a number representing what proportion of the data is to be used for testing purposes. This function then uses the above functions to process the text and convert a collection of text documents to a matrix of token counts of each lemmatized term. Terms which occur in the text less than five times are omitted. Lemmatized English stop words are also omitted as their relationship with the assigned class is negligible i.e. words such as 'in', 'of' or 'the' are unlikely to give any insight as to whether a review will be positive or negative so are omitted from training data. The scikit learn function 'train_test_split' is used to split the data into training and testing data in a ratio defined by the user. These training and testing subsets are then returned.

```python
def train_test_split_and_preprocess(df, perc_test, category):
    data=df.review #data required is the review column
    lemmatized_English_stopwords=lemmatize_tokens(text.ENGLISH_STOP_WORDS)
    #use scikit CountVectorizer,omitting English stopwords and terms with a frequency less than five, also reduce all words to their stems using lemma tokenizer
    vectorizer = CountVectorizer(stop_words=lemmatized_English_stopwords, min_df = 5, tokenizer=lemma_tokenizer)  
    data=vectorizer.fit_transform(data) #apply vectorizer to the review data
    display_common_terms(data, vectorizer, category) #display the top ten most common terms
    target=df['class'] #create the target, in this case the class of the review, i.e. whether it was positive or negative
    #use scikit learn train_test function to split the data into trainging and testing sets
    train_X, test_X, train_Y, test_Y=train_test_split(data, target, test_size=perc_test) 
    return train_X, train_Y, test_X, test_Y
```

The below function accepts training and testing subsets, three different classifiers are tested; Logistic Regression, Random Forest and Multinomial Naive Bayes. 

Logistic Regression is a statistical technique used to predict probability of binary response based on one or more independent variables. In this case the independent variables are the word counts of each term in a review. It is similar to linear regression model except used to predict a category as opposed to a numerical value and makes use of the logarithmic function to assign a binary classification.

Random Forest is an ensemble learning method used for classification. It operates by constructing a multitude of decision trees at training time and outputting the class that is the mode of the classes.

Multinomial Naive Bayes is suitable for classification with discrete features, it is a probabilistic classifier based on applying Bayes' theorem with naive independence assumptions between the features.

An Ensemble classifier is also created which combines the three aforementioned classifiers and designed around Condurcet's Jury Theorem, makes predictions based on the majority vote from the three classifers. According to Condurcet's Jury Theorem, if the probability of an individual voter making a correct vote is greater than 1/2, then adding more voters increases the probability that the majority decision is correct. In the limit, the probability that the majority votes correctly approaches 1 as the number of voters increases.

These four classifiers are compared against one another using five fold cross validation on the training data and the classifier with the highest average accuracy from each of the five folds is selected as the optimal classifier for that training data. The optimal classifier then makes predictions based on the unseen test data. These predictions, as well as the optimal model fitted to the training data and the name of the selected classifier are returned.

```python

def get_best_classifier(train_X, train_Y, test_X, test_Y):
    #initiate the classifiers
    LR = LogisticRegression(random_state=1, solver='lbfgs')
    RF = RandomForestClassifier(n_estimators=50, random_state=1)
    NB = MultinomialNB()
    Ens = VotingClassifier(
        estimators=[('lr', LR), ('rf', RF), ('gnb', NB)],
        voting='hard')
    results=[]
    clfs=[LR, RF, NB, Ens]
    labels=['Logistic Regression', 'Random Forest', 'Naive Bayes', 'Ensemble']
    print('We must now select a classifier')
    print('Using 5-fold cross validation, the scores evaluated for a select few classifiers are as follows:')
    #iterate through the classifiers and use an f1-score to evaluate their performance on each fold, print the average of these five scores for each classifier
    for clf, label in zip(clfs, labels):
        scores = cross_val_score(clf, train_X, train_Y, scoring='f1', cv=5)
        print("f1 Score: %0.2f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label))
        results.append(scores.mean())
    index_max=results.index(max(results)) #get the index of the highest averaged f1-score
    print('\nThe best performing classifier based on cross validation of training data is %s\n\n' %labels[index_max])
    best_classifier=clfs[index_max] #get the classifier associated with the highest f1 score
    best_classifier.fit(train_X, train_Y) #train the model on the training data provided using the best classifier
    predictions = best_classifier.predict(test_X) #use this model to make predictions on the previously unseen test data
    return best_classifier, predictions, labels[index_max] #return the model, the predictions on the test data and the name of the best classifier
```

The below function evaluates the performance of a classifier based on the predictions it made on the class of each review vs the actual class the review was assigned. Four different evaluation metrics are presented; the accuracy score, the precision score, the recall score and the F1 score. The confusion matrix is also displayed.

Confusion Matrix: The upper left hand corner displays the number of True Negative values, i.e. the number of negative predictions made which had a negative label. The lower right hand corner displays the number of True Positive values. The lower left displays the number of false positives, i.e. those positive predictions made that were actually negative, and the bottom right displays the number of false negatives. 

The accuracy score is the number if true positives plus true negatives divided by the total number of predictions. It is one of the most common evaluators and represents the fraction of predict corrections,  however, it is a poor evaluator when there is a large majority class in the training data. For example, if we are trying to predict a very unlikely event, the accuracy score might be very high even if the event failed to be predicted due to the disproportionately large number of true negatives compared to true positives. We see from the data that for all three categories, the majority of reviews are positive, thus a different metric may be more appropriate.

Recall Score: fraction of actual positives predicted as positive = TP/(TP+FN) (This may also be applied to the negative class)

Precision Score: fraction of predicted positives that are actually positive = TP/(TP+FP) (This may also be applied to the negative class)

F1 score: The weighted average of the precision and recall = 2 * (precision * recall) / (precision + recall)

The F1 score weights precision and recall equally (though this can be adjusted if one particular score is of more interest than the other). It is less sensitive to a majority class than the accuracy score thus is the optimal evaluation metric for this data. From the perspective of a person creating a classification model for hotel, automotive or fashion reviews, being able to predict a negative review is probably more important than being able to predict a positive review, this is because negative reviews have the potential to tarnish the reputation of the establishment but also hold key information as to how establishments may improve their services. Therefore the f1-score for negative reviews is the evaluation metric that will be kept for further analysis.


```python

def evaluate(pred, test_Y): 
    #create a classification report showing the precision, recall and f1-score for both the negative and positive class
    class_report=pd.DataFrame(classification_report(test_Y, pred, labels=[-1,1], target_names=["negative","positive"],output_dict=True)).transpose()
    print(class_report)
    #get the f1_score for negative reviews
    f1_neg=class_report['f1-score']['negative']
    cm = confusion_matrix(test_Y, pred) #create a confusion matrix of the predictions vs actuals
    group_names = ['True Neg','False Pos','False Neg','True Pos']
    group_counts = ["{0:0.0f}".format(value) for value in
                    cm.flatten()]
    labels = [f"{v1}\n{v2}" for v1, v2 in
              zip(group_names,group_counts)]
    labels = np.asarray(labels).reshape(2,2)
    #use the seaborn package to create a heatmap from the confusion matrix
    sns.heatmap(cm, annot=labels, fmt='', cmap='Blues')
    plt.ylabel('true label')
    plt.xlabel('predicted label')
    print('\n\n')
    return round(f1_neg, 4)
```

A function is created below which takes in two category names and uses the global function to call the relevant objects for these two categories then uses the function defined above to evaluate the classification model chosen for the first category and trained on data from the first category to make predictions on test data from the second category. The resulting f1-score for negatively classed reviews is then returned which will be used later on. 

```python
def one_category_on_another(cat1, cat2):
    #get chosen model name for category 1
    classifier=globals()['%s_model_name' %cat1] 
    #use the globals function to retrieve the dataframe of reviews and classifications for category 1
    df1=globals()['%s_reviews' %cat1]
    #do the same for category 2
    df2=globals()['%s_reviews' %cat2]
    #hold back 20 percent of the data for testing purposes
    perc_test=0.2
    #get the f1 score for negative reviews that was evaluated for predictions made for test data in category 1 by the selected model 
    #trained to data from category 1
    f1_orig=globals()['%s_f1_neg' %cat1]
    #call the above function and get the f1 score for negative reviews that was evaluated for predictions made for test data in category 2 
    #by the selected model for category 1 and trained to data from category 1
    f1_new=Model_for_another_cat(classifier, df1, df2, perc_test, f1_orig)
    return f1_new
```
